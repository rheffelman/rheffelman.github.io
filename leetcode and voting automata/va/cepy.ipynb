```python
import random

# Original voting data
voting_data = [
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['A', 'D', 'E', 'C', 'B'],
    ['B', 'E', 'D', 'C', 'A'],
    ['B', 'E', 'D', 'C', 'A'],
    ['B', 'E', 'D', 'C', 'A'],
    ['B', 'E', 'D', 'C', 'A'],
    ['B', 'E', 'D', 'C', 'A'],
    ['B', 'E', 'D', 'C', 'A'],
    ['B', 'E', 'D', 'C', 'A'],
    ['B', 'E', 'D', 'C', 'A'],
    ['B', 'E', 'D', 'C', 'A'],
    ['B', 'E', 'D', 'C', 'A'],
    ['B', 'E', 'D', 'C', 'A'],
    ['B', 'E', 'D', 'C', 'A'],
    ['C', 'B', 'E', 'D', 'A'],
    ['C', 'B', 'E', 'D', 'A'],
    ['C', 'B', 'E', 'D', 'A'],
    ['C', 'B', 'E', 'D', 'A'],
    ['C', 'B', 'E', 'D', 'A'],
    ['C', 'B', 'E', 'D', 'A'],
    ['C', 'B', 'E', 'D', 'A'],
    ['C', 'B', 'E', 'D', 'A'],
    ['C', 'B', 'E', 'D', 'A'],
    ['D', 'C', 'E', 'B', 'A'],
    ['D', 'C', 'E', 'B', 'A'],
    ['D', 'C', 'E', 'B', 'A'],
    ['D', 'C', 'E', 'B', 'A'],
    ['D', 'C', 'E', 'B', 'A'],
    ['D', 'C', 'E', 'B', 'A'],
    ['D', 'C', 'E', 'B', 'A'],
    ['D', 'C', 'E', 'B', 'A'],
    ['D', 'C', 'E', 'B', 'A'],
    ['E', 'B', 'D', 'C', 'A'],
    ['E', 'B', 'D', 'C', 'A'],
    ['E', 'B', 'D', 'C', 'A'],
    ['E', 'B', 'D', 'C', 'A'],
    ['E', 'B', 'D', 'C', 'A'],
    ['E', 'B', 'D', 'C', 'A']
]

# Function to generate unique ballots for each pixel
def extend_voting_data(original_data):
    extended_data = []
    for ballot in original_data:
        # Generate a random permutation of the original ballot
        random.shuffle(ballot)
        extended_data.append(ballot)
    return extended_data

# Extend the original voting data
extended_voting_data = extend_voting_data(voting_data)

# Output file name
output_file = "extended_voting_data.txt"

# Write extended voting data to a text file
with open(output_file, 'w') as file:
    for idx, ballot in enumerate(extended_voting_data):
        file.write(f"Pixel {idx}: {','.join(ballot)}\n")

print(f"Extended voting data written to {output_file}")
```
